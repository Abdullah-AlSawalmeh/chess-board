```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from random import random
```

```python
class ChessBoard:
    def __init__(self):
        self.cube = np.zeros((8,8,3))
        self.first_render()
        self.red = None
        self.blue = None

    def first_render(self):
        for i , row in enumerate(self.cube):
            for j,column in enumerate(row):
                for k , channel in enumerate(column):
                    if i % 2 == 0:
                        if j % 2 == 0:
                            column[k] = 1
                    if i % 2 != 0:
                        if j % 2 != 0:
                            column[k] = 1                   

        plt.imshow(self.cube)

    def render(self):
        plt.imshow(self.cube)


    def add_red(self,row,column):
        self.cube[row][column][0] = 1
        self.cube[row][column][1] = 0.2
        self.cube[row][column][2] = 0
        self.red = [row,column]

    def add_blue(self,row,column):
        self.cube[row][column][0] = 0
        self.cube[row][column][1] = 1
        self.cube[row][column][2] = 1
        self.blue = [row,column]
    
    def is_under_attack(self):
        if self.red[0] == self.blue[0]:
            # print("On same row")
            return True
        elif self.red[1] == self.blue[1]:
            # print("On same column")
            return True
        elif (self.red[0]+1)/(self.red[1]+1) == 1 and (self.blue[0]+1)/(self.blue[1]+1) == 1:
            # print("On same diagonal")
            return True
        else:
            return False
```

```python
# Test if on the same column
game = ChessBoard()
game.add_red(1,5)
game.add_blue(5,5)
game.render()
game.is_under_attack()
assert game.is_under_attack() == True
```

```python
# Test if on the same row
game = ChessBoard()
game.add_red(5,3)
game.add_blue(5,2)
game.render()
game.is_under_attack()
assert game.is_under_attack() == True
```

```python
# Test if on the same diagonal
game = ChessBoard()
game.add_red(4,4)
game.add_blue(0,0)
game.render()
game.is_under_attack()
assert game.is_under_attack() == True
```

```python
# Test if Not under Attack
game = ChessBoard()
game.add_red(1,4)
game.add_blue(3,5)
game.render()
game.is_under_attack()
assert game.is_under_attack() == False
```
